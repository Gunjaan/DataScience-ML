### <center> Predicting the survival rate of a passenger in Titanic

```python
import pandas as pd
import numpy as np
DF = pd.read_csv("titanic.csv")
DF.head()
```

```python
df= DF
df.shape
```

```python
df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'], axis='columns')
df.head(3)
```

```python
df.isna().sum() # sum of all unavailable data in a column
```

```python
avgAge = df.Age.mean()
avgAge=int(avgAge)
# replacing na values with mean
df['Age'].fillna(value=avgAge, inplace=True)
```

```python
df.isna().sum()
```

```python
outputDF = DF.Survived
outputDF.head(3)
```

```python
import sklearn
# from sklearn.model_selection import split_and_test
```

```python
df.head()
```

```python
# since 'sex' is not an integer, we can not fit out model just yet.
# Converting it into a quantitative variable:
dummies= pd.get_dummies(df["Sex"])
dummies
```

```python
df= pd.concat([df, dummies], axis='columns')
df
```

```python
df=df.drop(['Sex'], axis="columns")
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(df,outputDF,test_size=0.2) 
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(xTrain,yTrain)
```

```python
print("Accuracy of model:", end=" ") 
print(model.score(xTest, yTest)*100,"%")
```
