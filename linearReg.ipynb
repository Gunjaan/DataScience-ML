```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
from word2number import w2n
```

```python
df=pd.read_csv("/content/testt - Sheet1.csv")
df
```

```python
medianTestScore = math.floor(df['test_score(out_of _10)'].mean())
df['test_score(out_of _10)']=df['test_score(out_of _10)'].fillna(medianTestScore)
df
```

```python
df['experience']= df['experience'].fillna("zero")
df
```

```python
df.experience = df.experience.astype(str)
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out_of _10)','interview_score(out_of_10)']],df['salary'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
print("The predicted salary is: ", reg.predict([[2, 9, 6]]))

```
