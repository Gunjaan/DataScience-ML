## **QUESTION**
**Use famous iris flower dataset from sklearn.datasets to predict flower species using random forest classifier.**

**Measure prediction score using default n_estimators (10)
Now fine tune your model by changing number of trees in your classifer and tell me what best score you can get using how many trees**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print(dir(iris))
```

```python
import pandas as pd
df=pd.DataFrame(iris.data)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

---
**Training and the model and prediction**

```python
x = df.drop('target',axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

<center> <b>increaing n_estimator value by 10

```python
# n_estimators : int, default=100 
# Meaning -> The number of trees in the forest
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20, )
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

## **Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted=model.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
