```python
import pandas as pd
df = pd.read_csv("carprices.csv")
df
```

```python
dummies= pd.get_dummies(df["Car Model"])
dummies
```

```python
mergeddf= pd.concat([df, dummies], axis='columns')
mergeddf
```

```python
finaldf = mergeddf.drop(['Car Model', 'Mercedez Benz C class'], axis="columns")
finaldf
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(finaldf[['Mileage',	'Age(yrs)',	'Audi A5',	'BMW X5']],df['Sell Price($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

**Price of mercedez benz that is 4 year old with mileage 45000**

```python
reg.predict([[45000,4,0,0]])
```

**Price of BMW X5 that is 7 year old with mileage 86000**

```python
print("The predicted price is: ", reg.predict([[86000, 7, 0, 1]]))
```

```python
a = finaldf.drop('Sell Price($)',axis='columns')
b = finaldf['Sell Price($)']
reg.score(a,b)
```
